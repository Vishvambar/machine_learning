```python
import numpy as np 
```

```python
contestant_dtype = [('Name', 'U10'), ('Score','i4'), ('Time', 'f4')]
```

```python
participants = np.array([
    ('Rohan',  120, 45.5),
    ('Priya',  300, 20.1),  # Winner candidate
    ('Amit',   150, 50.0),
    ('Sneha',  280, 25.0),  # Winner candidate
    ('Vikram', 50,  60.0),
    ('Rahul',  290, 22.5),  # Winner candidate
    ('Anjali', 200, 40.0)
], dtype=contestant_dtype)
```

```python
scores = participants['Score']
```

```python
print(scores)
```

```python
k = 3
```

```python
partition = np.argpartition(scores, k)
print("partiotion indices : ",  partition)
print("Partiotion Scores : ", scores[partition])
```

```python
top3= partition[-k:]
print("Top3 indices : ", top3)
print ("top 3 scores : " , scores[top3])
```

```python
participants_top3 = participants[top3]
print("Top 3 participants before sorting:\n", participants_top3)
```
