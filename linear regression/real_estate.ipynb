```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

n = 100  # 100 houses

# Features
area = np.random.randint(400, 2500, n)       # sqft
bedrooms = np.random.randint(1, 5, n)        # 1 to 4
age = np.random.randint(0, 30, n)             # years

# Hidden (unknown) pricing rule
price = (
    area * 0.04 +        # price per sqft
    bedrooms * 8 -       # bedroom value
    age * 0.6 +          # depreciation
    np.random.randn(n) * 5  # noise
)

# Create DataFrame
df = pd.DataFrame({
    "Area_sqft": area,
    "Bedrooms": bedrooms,
    "Age_years": age,
    "Price_lakhs": price
})

print(df.head())
df.to_csv("house_prices.csv", index=False)


x = df['Area_sqft'].values
y = df['Price_lakhs'].values

# Normalize Features 

x = (x - x.mean())/x.std()


w = 0.0 
b = 0.0 

learning_rate = 0.01
epochs = 3000 
n = len(x)


for epoch in range(epochs):
    y_pred = w*x+b

    dw = (2/n)*np.sum((y_pred-y)*x)
    db = (2/n)*np.sum((y_pred-y))

    w = w - learning_rate*dw 
    b = b - learning_rate*db

    loss = np.mean((y_pred - y) ** 2)
    print(f"Epoch {epoch} | Loss: {loss:.2f} w : {w:.4f} b: {b:.4f}")



new_area = 1399
new_area_norm = (new_area-df["Area_sqft"].mean())  / df["Area_sqft"].std()

predicted_price = w*new_area_norm + b

print(f"Predicted price for {new_area} sqft house  {predicted_price} lakhs")


```
