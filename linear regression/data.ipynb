```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# STEP 1: Generate Data
# -------------------------------

np.random.seed(42)

X = np.random.rand(50) * 10          # 50 input values (0 to 10)
noise = np.random.randn(50) * 2
y = 2 * X + 5 + noise                # Unknown rule to the model

# -------------------------------
# STEP 2: Initialize Dumb Model
# -------------------------------

w = 0.0   # weight (slope)
b = 0.0   # bias (intercept)

# Prediction function
def predict(x, w, b):
    return w * x + b

# -------------------------------
# STEP 3: Loss Function (MSE)
# -------------------------------

def loss_function(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# -------------------------------
# STEP 4 & 5: Training (Gradient Descent)
# -------------------------------

learning_rate = 0.01
epochs = 10000
n = len(X)

losses = []

for epoch in range(epochs):
    # Predictions on all data
    y_pred = predict(X, w, b)

    # Loss
    loss = loss_function(y, y_pred)
    losses.append(loss)

    # Gradients
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss:.4f} | w: {w:.4f} | b: {b:.4f}")

# -------------------------------
# STEP 6: Test the Model (Inference)
# -------------------------------

x_test = 7.0
y_test_pred = predict(x_test, w, b)

print("\n--- TEST RESULT ---")
print(f"For x = {x_test}, predicted y = {y_test_pred:.2f}")

# -------------------------------
# STEP 7: Visualization
# -------------------------------

# Plot loss curve
plt.figure(figsize=(7, 4))
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Over Time")
plt.show()

# Plot data and learned line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Training Data")
plt.plot(X, predict(X, w, b), color="red", label="Learned Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Learned From Scratch")
plt.legend()
plt.show()
```
