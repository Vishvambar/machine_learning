```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
# Using Linear Regression

df = pd.DataFrame({
    'Experience' : [1, 2, 5],
    'Salary' : [30000, 35000, 80000]
})

x = df['Experience'].values
y = df['Salary'].values


w=0.0
b=0.0
learning_rate = 0.001 
epochs = 1000
n = len(x)

for epoch in range(epochs):
    y_pred = w * x + b
    error = y_pred - y
    dw = (2/n) * np.sum(error * x)
    db = (2/n) * np.sum(error)
    w = w - learning_rate * dw
    b = b - learning_rate * db
    if epoch % 100 == 0:
        loss = np.mean(error ** 2)
        print(f"Epoch {epoch}, Loss: {loss:.2f}, w: {w:.2f}, b: {b:.2f}")


```

```python
# using polynomial regression

df = pd.DataFrame({
  'Experience' : [1, 2, 5],
  'Salary'  : [30000, 35000, 80000]
})

x= df['Experience'].values
x2 = x** 2
y = df['Salary'].values

# X = np.column_stack((x1, x2))

w1 = 0.0 
w2 = 0.0
b = 0.0
learning_rate = 0.001
epochs = 1000
n = len(x)

for epoch in range(epochs):
    y_pred = w1* x + w2* x2 + b
    error = y_pred - y


    dw1 = (2/n) * np.sum(error * (x))
    dw2 = (2/n) * np.sum(error * (x2))
    db = (2/n) * np.sum(error)


    w1 = w1 - learning_rate * dw1
    w2 = w2 - learning_rate * dw2
    b = b - learning_rate * db

    if epoch % 100 == 0:
      loss = np.mean(error ** 2)
      print(f"Epoch {epoch} | Loss: {loss:.2f} | w1: {w1:.2f}, w2: {w2:.2f}, b: {b:.2f}")

    
```

```python
# Linear regression vs Polynomial regression visualization
plt.scatter(x, y, color='blue', label='Data Points')
x_range = np.linspace(0, 6, 100)
y_linear = w * x_range + b
y_poly = w1 * x_range + w2 * (x_range ** 2) + b
plt.plot(x_range, y_linear, color='red', label='Linear Regression')
plt.plot(x_range, y_poly, color='green', label='Polynomial Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Linear vs Polynomial Regression')
plt.legend()
plt.show()

```
