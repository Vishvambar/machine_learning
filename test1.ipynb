The Scenario: You are managing inventory. You want to identify which product is "stagnant". A product is "stagnant" if it has High Stock (> 50) but Low Sales Rate (< 2 items/day). Among these stagnant products, you want to find the one with the Highest Price so you can put it on a clearance sale first.

```python
import numpy as np

items = np.array(["Phone", "Laptop", "Mouse", "Keyboard", "Monitor", "Webcam"])
stock = np.array([10,      60,       55,      120,        5,         70])
sales = np.array([5,       1.5,      10,      0.5,        0.2,       1.0])
prices = np.array([500,    1000,     20,      50,         200,       80])
```

```python
mask = (stock >50) & (sales <2)
```

```python
filtered_item = items[mask]
filtered_stock = stock[mask]
filtered_price = prices[mask]
filtered_sales = sales[mask]
```

```python
highest_price = np.argmax(filtered_price)
```

```python
item_name = filtered_item[highest_price]
print(item_name)
```
