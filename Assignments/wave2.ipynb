```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]
```

```python
prices = [100, 102, 98, 90, 85]
```

```python
plt.plot(days, prices)
plt.axhline(y = 95 , color = 'red' , linestyle = '--' , label = 'Buy Price Threshold')
plt.xlabel('Days')
plt.ylabel('Prices')
plt.legend()
plt.show()
```
