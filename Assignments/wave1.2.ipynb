Scenario: DealConnect. Filter products that are "Good Deals". Definition of Good Deal: (Price < 100) OR (Rating > 4.5).

```python
import numpy as np 
import pandas as pd
```

```python
prices = np.array([50,200,1000,20,80])
ratings = np.array([3.0, 4.8, 4.9, 2.5, 4.6])
products = np.array(['A','B','C','D','E'])
```

```python
good_deal_mask = (prices < 100) | (ratings > 4.5)
good_deal_mask
```

```python
good_deals=products[good_deal_mask]
print ("the good products to buy are : ", good_deals)
```
