```python
import numpy as np 
```

```python
# --- 1. THE DATA (The Past) ---
# Features: [Action_Amount, Romance_Amount]
movie_a_features = np.array([5.0,0.0])
movie_b_features = np.array([0.0,5.0])
```

```python
# Your Ratings (Y_true)
# You gave Mad Max 5 stars, The Notebook 1 star.
y_true_a = 5.0 
y_true_b = 1.0
```

```python
# User Taste - the computer dont know your test yet it guess some random
learning_rate =0.01
user_taste = np.array([0.5,0.5])
```

```python
# Training loop
print(f"for initial random taste \n Action :  {user_taste[0]:.2f} Romance : {user_taste[1]:.2f}")

for i in range(1000):

    #movie a madmax 
    pred_a = np.dot(user_taste , movie_a_features)
    error_a = pred_a - y_true_a
    slope_a = error_a * movie_a_features
    user_taste = user_taste - learning_rate*slope_a

    #movie b notebook 
    pred_b = np.dot(user_taste , movie_b_features)
    error_b = pred_b - y_true_b 
    slope_b = error_b * movie_b_features
    user_taste = user_taste - learning_rate * slope_b 

print('--'*30)
print (f'after training for 100 iterations \n Action :  {user_taste[0]:.2f} Romance : {user_taste[1]:.2f}')
```

```python
import numpy as np 

# you love action 5.0 and hate romance -2.0 

true_user_tastes = np.array([5.0 , -2.0])

# we have 100 random movies 
#Each movie has random action romance levels between 0 to 5 

num_movies = 1000
movie_features_list = np.random.rand(num_movies , 2 ) * 5

# generate your true ratings for these movies  ( plus some noise or mood swings )
# this simulates you actually watching them 

rating_list = []
for movie in movie_features_list: 
    # ratings = (action*5.0)+(romance*(-2.0))
    true_rating = np.dot(true_user_tastes , movie)
    rating_list.append(true_rating)

# Learning phase 
# the model starts stupid guess 

model_taste = np.array([0.0,0.0])
learning_rate  = 0.1

print(f"Start taste : {model_taste}")

# we loop through movie one by one 

for i in range(num_movies):

    # pick one single movie and its rating 
    features = movie_features_list[i]
    actual_rating = rating_list[i]

    # Make prediction 
    prediction = np.dot(model_taste , features)

    # calculate error for this one movie 
    error = prediction - actual_rating

    # update weights immediately 
    model_taste = model_taste - learning_rate * (error * features )

    # check in every 200 movies 
    if i % 200 == 0 :
        print(f"Movie #{i} : Learned Taste ={model_taste}")

print("-"*30)
print ("Final Learned Taste : ", model_taste)
print (f"True User taste : {true_user_tastes}" )
```
