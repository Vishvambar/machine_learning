```python
import numpy as np
```

```python
# Glucose , BP , Age
Data = np.array([
    [100, 80, 25],
    [120, 140, 60], 
    [90, 70, 30], 
    [200, 150, 55], 
    [80, 60, 20]])
```

```python
print("Mean of Glucose : " , np.mean(Data[:,0]))
print("The standard Deviation of Glucose : " , np.std(Data[:,0]))
print("Mean of Blood Pressure : " , np.mean(Data[:,1]))
print("The standard Deviation of Blood Pressure : " , np.std(Data[:,1]))
print("Mean of Age : " , np.mean(Data[:,2]))
print("The standard Deviation of Age : " , np.std(Data[:,2]))
mean_Data = np.mean(Data, axis=0)
std_data = np.std(Data, axis=0)

```

```python
Normalised_Data = (Data - mean_Data) / std_data
```
