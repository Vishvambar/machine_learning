```python
import numpy as np 

def log_loss(y , y_hat):
    eps = 1e-15   #avoid log 0 
    y_hat = np.clip(y_hat , eps ,1-eps)
    return -np.mean(y*np.log(y_hat)+(1-y)* np.log(1-y_hat))
```

```python
y = np.array([1,0])
y_hat = np.array([0.9 , 0.1])
log_loss(y , y_hat)
```

```python
y     = np.array([1, 0, 1, 0])
y_hat = np.array([0.95, 0.90, 0.60, 0.10])

log_loss(y, y_hat)
```
