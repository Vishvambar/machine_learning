```python
import numpy as np 
```

```python

# input features - hours studied 
x = np.array([1,2,3,4,5]).reshape(-1,1)
# Labels : 0-fail , 1-pass
y = np.array([0,0,0,1,1]).reshape(-1,1)

m = x.shape[0]
```

```python
#Sigmoid function 

# converts the raw scores in probablity 

def sigmoid (z):
    return 1/(1+np.exp(-z))
```

```python
# Loss log 

def log_loss(y , y_hat):
    eps  = 1e-15
    y_hat = np.clip(y_hat , eps , 1-eps )
    return -np.mean(y*np.log(y_hat)+(1-y)*np.log(1-y_hat))
```

```python
# forward Pass - linear + sigmoid 

def forward (x , w, b):
    z = np.dot (x, w)+b 
    y_hat = sigmoid (z)
    return y_hat
```

```python
w = np.zeros((1,1))
b = 0.0 
learning_rate = 0.1 
epochs = 1000
```

```python
for epoch in range(epochs):
    y_pred = np.dot(x, w)+b      #Linear
    
    y_hat = sigmoid(y_pred)      #Probability 
 
    loss = log_loss(y , y_hat)   # loss 

    #  Backward Pass Manual (Gradient Descent)
    dy = y_hat - y 

    dw = (1/m) * np.dot(x.T , dy)
    db = (1/m) * np.sum(dy)

    # Update
    w = w-learning_rate*dw
    b = b-learning_rate*db

    # Logging 
    if epoch % 100 ==0:
        print(f"Epoch {epoch} : Loss {loss} , w {w[0][0]} , b {b}")
```

```python
new_hours = np.array([[2.5], [4.5]])

z = np.dot(new_hours, w) + b
probs = sigmoid(z)

print("\nPredictions:")
for h, p in zip(new_hours, probs):
    print(f"Hours studied: {h[0]} → Pass probability: {p[0]:.2f}")

```
