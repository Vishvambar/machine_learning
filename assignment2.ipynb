```python
import numpy as np
dirty_data = np.array([100, 50, np.nan, 45, 300, np.nan, 600])
```

```python
dirty_data.mean()
```

```python
clean_data = np.where(np.isnan(dirty_data), 0 ,dirty_data)
print (f"{clean_data} , {clean_data.mean()}")
```

```python
mean_value = np.nanmean(dirty_data)
print(mean_value)
```
