```python
input = 10
weight = 0.5 
target = 20 
learning_rate = 0.01
```

```python
print(f"Start: Input={input }, Weight={weight}, Prediction={input * weight}")
```

```python
for i in range(20):
    prediction = input * weight 
    error = prediction - target 
    slope = error * input 
    weight = weight - (slope * learning_rate)
    print(f"Iter {i+1}: Prediction={prediction:.2f} | Error={error:.2f} | New Weight={weight:.2f}")

print (f"Final Weight : {weight:.2f}")
```

```python
input_data = 10
target_answer = 20
weight = 0.5 
learning_rate = 0.01 

# We start with a massive error just to get into the loop
error = 100 
iteration = 0

# LOOP UNTIL ERROR IS TINY (Close to 0)
# abs(error) removes the negative sign
while abs(error) > 0.001: 
    iteration += 1
    
    prediction = input_data * weight
    error = prediction - target_answer  
    
    slope = error * input_data
    weight = weight - (slope * learning_rate)
    
    print(f"Iter {iteration}: Prediction= {prediction:.4f} Weight={weight:.4f}, Error={error:.4f}")
    
    # Safety break: Stop if it takes too long
    if iteration > 1000:
        print("Took too long! Stopping.")
        break

print(f"Done! It took {iteration} steps to learn.")
```

```python
import numpy as np

# 1. The Setup (Our Data)
# Let's use 5 data points to make it stable
X = np.array([1, 2, 3, 4, 5])
# The True Answer key: y = 2x + 10
Y_true = np.array([12, 14, 16, 18, 20])

# 2. Initialization (Random Guess)
w = 0.0  # Start slope at 0
b = 0.0  # Start intercept at 0
learning_rate = 0.01
epochs = 6000 # Number of loops

# 3. The Training Loop
for i in range(epochs):
    
    # --- A. FORWARD PASS (Prediction) ---
    # We predict for ALL 5 points at once
    Y_pred = w * X + b
    
    # --- B. MEASURE ERROR ---
    # "Mean Squared Error" logic, but simplified for gradient calc
    error = Y_pred - Y_true
    
    # --- C. CALCULATE GRADIENTS (The Direction) ---
    # Gradient for w: Average of (error * x)
    w_grad = np.mean(error * X)
    
    # Gradient for b: Average of (error)
    b_grad = np.mean(error)
    
    # --- D. UPDATE WEIGHTS (The Step) ---
    w = w - (learning_rate * w_grad)
    b = b - (learning_rate * b_grad)
    
    # Print progress every 100 steps
    if i % 100 == 0:
        # Calculate MSE just for display
        mse = np.mean(error**2)
        print(f"Epoch {i}: w={w:.3f}, b={b:.3f}, Error={mse:.4f}")

print("-" * 30)
print(f"FINAL RESULT: w={w:.3f}, b={b:.3f}")
print(f"TARGET:       w=2.000, b=10.000")



```

```python
print("-" * 30)
print("THE REAL TEST (Predicting the Unknown)")

# The model has NEVER seen the number 100.
new_input = 100 

# But it knows the RULE (w=2.0, b=10.0)
prediction = w * new_input + b

print(f"Input: {new_input}")
print(f"Prediction: {prediction:.2f}") 
# Output will be approx 210.00
```
