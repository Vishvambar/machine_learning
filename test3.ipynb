Scenario : 
You scraped data from two sources. You need to combine them into a single dataset, but you must track where each item came from.

```python
import numpy as np

# [Price, Rating]
amazon_data = np.array([
    [1000, 4.5],
    [500,  4.2],
    [200,  3.9]
]) # Shape (3, 2)

flipkart_data = np.array([
    [950,  4.6],
    [480,  4.1]
]) # Shape (2, 2)
```

```python
amazon_id = np.zeros((3, 1))
flipkart_id = np.ones((2,1))
```

```python
amazon_final = np.hstack((amazon_data, amazon_id))
flipkart_final = np.hstack((flipkart_data , flipkart_id))
```

```python
final_data = np.vstack((amazon_final , flipkart_final))
```
