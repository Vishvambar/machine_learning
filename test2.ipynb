You need to identify patients who are in "Shock". Definition of Shock: (Temp < 98) OR (Heart Rate > 120).

```python
import numpy as np
```

```python
# [Temp, HR]
vitals = np.array([
    [98.6, 72],   # Normal
    [97.0, 110],  # Shock (Low Temp)
    [102.0, 130], # Shock (High HR)
    [96.5, 60],   # Shock (Low Temp)
    [99.0, 80]    # Normal
])
```

```python
shock_mask = (vitals[: , 0]< 98.0) | (vitals[:, 1]>120)
```

```python
shock_patients = vitals[shock_mask]
```

```python
mean_hr = shock_patients[: , 1].mean()
mean_hr
```
