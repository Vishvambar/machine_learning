Situation (Very Realistic)

You are monitoring a patientâ€™s heart rate every minute in the ICU.

Heart rate naturally fluctuates

Doctors donâ€™t react to every fluctuation

They care about:

Baseline trend â†’ Is HR generally increasing?

Stability â†’ Is HR becoming erratic?

ðŸ‘‰ This is exactly where
rolling mean + rolling std are used.

```python
import pandas as pd

hr_data = [72, 74, 73, 75, 74, 120, 130, 125, 122, 118]

df = pd.DataFrame({
    "Minute": range(len(hr_data)),
    "Heart_Rate": hr_data
})

df

```

```python
df["HR_mean_3"] = df["Heart_Rate"].rolling(3).mean()
```

```python
df
```

```python
df["HR_std_3"]= df["Heart_Rate"].rolling(3).std()
```

```python
df
```

```python
df["alert"] = df["Heart_Rate"] > (df["HR_mean_3"]+2 * df["HR_std_3"])
```
