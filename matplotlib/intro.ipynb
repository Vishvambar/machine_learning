```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y ,color ='green' , linestyle = '--' , marker = 'o')
plt.title("Patient Recovery Rate")
plt.xlabel("Days")
plt.ylabel("Immunity Score")
plt.grid(True)
plt.axhline(y=22, color ='red' , label = "critical threshold")
plt.legend()
```

```python
Time = [8,12,16,20]
Temp = [98.6, 99.5, 102.0, 100.5]
```

```python
plt.plot(Time, Temp, color = 'blue' , marker='o')
plt.title("Patient fever Monitor")
plt.axhline(y=100 ,   color ='red' ,label = "Fever Threshold" )
plt.legend()
plt.show()
```
