Scenario: Find patients with (Heart Rate > 100) AND (Oxygen < 90).

```python
import numpy as np

# Columns: [Patient_ID, Heart_Rate, Oxygen_Level]
# ID is just a number here for simplicity
patient_data = np.array([
    [101, 72, 98],   # Healthy
    [102, 115, 90],  # High HR, but Oxygen OK
    [103, 60, 99],   # Healthy
    [104, 120, 85],  # CRITICAL: High HR + Low Oxygen
    [105, 55, 88],   # CRITICAL: Low HR + Low Oxygen
    [106, 250, 98]   # SENSOR ERROR (Heart rate 250? Unlikely)
])
```

```python
heart_rate = patient_data[: , 1]
heart_rate
```

```python
oxygen = patient_data[: , 2]
oxygen
```

```python
masking = (heart_rate > 100 ) & (oxygen < 90)
```

```python
print(f"Super Critical patient :  {patient_data[masking]}")
```

Data Cleaning for negative values

```python
products = np.array([
    [500, 4.5],
    [0, 3.2],     # ERROR: Price 0
    [1200, 4.0],
    [-99, 1.0],   # ERROR: Negative Price
    [300, 4.8]
])
```

```python
prices = products[:, 0]
```

```python
prices
```

```python
valid_mask = prices >0
```

```python
valid_mask
```

```python
cleaned_data = products[valid_mask]
```

```python
print(f"Cleaned Deal Data : {cleaned_data}")
```
