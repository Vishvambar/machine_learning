```python
import numpy as np
```

```python
products = np.array(["Mouse", "Laptop", "Monitor", "Cable", "Headphones"])
prices =   np.array([20,      1000,     200,       10,      50])
ratings =  np.array([3.5,     4.8,      4.2,       3.9,     4.5])
```

```python
rating_mask = ratings > 4 
```

```python
filtered_prices = prices[rating_mask]
filtered_products = products[rating_mask]
```

```python
filtered_prices
```

```python
filtered_products
```

```python
min_index = np.argmin(filtered_prices)
```

```python
print(filtered_products[min_index])
```
