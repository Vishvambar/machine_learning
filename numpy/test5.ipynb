```python
import numpy as np

players_dtype = [('Player','U10'),('Runs','i4')]

players = np.array([
    ('Ravi',  40),
    ('Neha',  120),
    ('Aman',  75),
    ('Isha',  200),
    ('Kabir', 30),
    ('Simran',160),
    ('Farah', 95)
], dtype=players_dtype)

print(players)
```

```python
scores = players['Runs']
```

Using the K partiotion so we get the first 2 smallest on one side and on the other side we get the biggest 2 on the other side.


```python
k  = 2
```

```python
idx = np.argpartition(scores, k)
```

```python
print("Partiotion indices : ", idx)

print("Scores with that order : ", scores[idx])

```

```python
top2_using_k = idx[-2:]
print("Player Top 2 : " , players[top2_using_k])
```

```python
idx2 = np.argpartition(scores, -k )
```

```python
print ("Partiotion with -k indices ", idx2)
print ("scores with the -k partiotion : ", scores[idx2])
```

```python
top2_using_negk = idx[-2:]
```

```python
print("Player Top 2 using -k : " , players[top2_using_negk])
```
