```python
import numpy as np 
```

```python
# [WBC , TEMP , HEART RATE]

patient_data = np.array([
    [5000, 98.6, 72],
    [11000, 102.4, 110],  # Sick patient
    [4500, 97.5, 65],
    [7000, 99.0, 80],
    [8500, 100.1, 85]
    ])
```

```python
print(f"original shape : {patient_data.shape}")
```

```python
feature_means = patient_data.mean(axis=0)
```

```python
print(f"Means : {feature_means}")
```

```python
print(f"Shape of Means : {feature_means.shape}")
```

```python
centered_data = patient_data- feature_means
```

```python
print ("centered data : ", centered_data)
```

```python
print(f"\nNew Means (should be ~0): {centered_data.mean(axis=0)}")
```

```python
std_patient_data = patient_data.std(axis=0)
```

```python
print(std_patient_data)
```

```python
dividation = centered_data/std_patient_data
```

```python
print("\nStandardized data : \n", dividation)
```

```python
import numpy as np

clinical_data = np.array([
    [[72, 98], [70, 99], [71, 98]],
    [[110, 92], [115, 90], [120, 88]]
])

# Step 1 â€” Max subtraction (Numerical Stabilization)
max_vals = clinical_data.max(axis=2, keepdims=True)
result = clinical_data - max_vals        # shifted values
```

```python
max_vals
```

```python
clinical_data
```

```python
result
```

```python
exp_val = np.exp(result)
exp_val
```

```python
sum_exp = exp_val.sum(axis=2 , keepdims=True)
sum_exp
```

```python
softmax_output = exp_val / sum_exp
```
