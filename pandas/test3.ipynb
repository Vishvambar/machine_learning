```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Name ": ["John", "Alice", "Bob", "Charlie", "Eve"],
    "Age": [45, 30, np.nan, 50 , np.nan],
    "Sugar_Level": [120, np.nan, 90, 110, np.nan]
})
df
```

```python
df  = df.dropna(subset=['Age'])
df
```

```python
mean_sugar = df["Sugar_Level"].mean()
df["Sugar_Level"] = df["Sugar_Level"].fillna(mean_sugar)

# Final Output
print("Cleaned Data:\n", df)
```
