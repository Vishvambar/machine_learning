```python
import pandas as pd 
```

```python

# Table 1: Who are they?
patients = pd.DataFrame({
    "Patient_ID": [101, 102, 103],
    "Name": ["Alice", "Bob", "Charlie"],
    "Blood_Type": ["A+", "O-", "B+"]
})

# Table 2: What is wrong with them?
labs = pd.DataFrame({
    "Patient_ID": [101, 103, 104],  # Note: 102 is missing, 104 is new!
    "Infection_Score": [85, 90, 55]
})
```

```python
# Link them on the common column "Patient_ID"
merged_df = pd.merge(patients, labs, on="Patient_ID", how="inner")
print(merged_df)
```
