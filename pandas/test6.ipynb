```python
import numpy as numpy
import pandas as pd
```

```python
Products  = pd.DataFrame({
    "ID" : [1, 2, 3, 4],
    "Name": ["Mouse", "Keyboard", "Laptop", "Monitor"]

})

Discounts = pd.DataFrame({
    "ID": [2,3,5],
    "Discount": [0.10,0.20,0.50]
})
```

```python
Products
```

```python
Discounts
```

```python
# Link them on the common column "Patient_ID"
joined_data = pd.merge(Products, Discounts, on="ID", how="left")
print(joined_data)
```

```python
cleaned_data = joined_data.fillna(0)
```
