```python
import pandas as pd
import numpy as np

data = {
    "Temperature": [98.6, np.nan, np.nan, 99.1, np.nan]
}

df = pd.DataFrame(data)
print(df)
```

```python
df = df.ffill()
```

```python
df
```

```python
df = pd.DataFrame({
    "Time": pd.date_range("2025-01-01 10:00", periods=5, freq="T"),
    "Heart_Rate": [72, np.nan, 75, np.nan, np.nan]
})

df.set_index("Time", inplace=True)
print(df)
```

```python
df.ffill(inplace=True)
print(df)
```
