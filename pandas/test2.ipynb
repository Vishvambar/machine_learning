```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Patient": ["A", "B", "C", "D"],
    "HR": [72, np.nan, 80, 65],       # Patient B is missing Heart Rate
    "Temp": [98.6, 99.0, np.nan, np.nan] # C and D are missing Temp
})

print(df)
```

```python
# finding the NULL or missing values 
print(df.isna().sum())
```

```python
# The dangerous method is to drop the rows with missing values
# df = df.dropna()
# print(df)

# The safer method is to fill the missing values with a specific value
df = df.fillna(0)
print(df)
```
