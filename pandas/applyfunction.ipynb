```python
df = pd.DataFrame({
    'Name': ['John', 'Alice', 'Bob'],
    'Age': [25, 30, 35]
})
```

```python
df
```

```python
def age_category(age):
    if age < 30:
        return 'Young'
    else:
        return 'Old'
```

```python
df["Category"] = df["Age"].apply(age_category)
```

```python
df
```

```python
df = pd.DataFrame({
    "Heart_Rate": [98, 105, 120],
    "Oxygen": [96, 95, 92]
})
```

```python
df
```

```python
def Risk(row):
    if row['Heart_Rate'] > 100 and row['Oxygen'] < 95:
        return 'High Risk'
    else:
        return 'Normal'
```

```python
df['Risk_Level']= df.apply(Risk, axis=1)
```

```python
df
```
