```python
import pandas as pd 
```

```python

# 1. Raw Data (Strings)
data = {
    "Time": ["2025-01-01 08:00", "2025-01-01 08:15", "2025-01-01 08:30", "2025-01-01 08:45",  "2025-01-01 09:00"],
    "HR": [70, 72, 75,80, 110] # Spiking at 9 AM
}
df = pd.DataFrame(data)
df
```

```python
# 2. The Conversion (String -> DateTime)
df["Time"] = pd.to_datetime(df["Time"])
df
```

```python
df
```

```python

# 3. The Indexing (Time MUST be the index for resampling)
df = df.set_index("Time")
df
```

```python

# 4. The Magic: Resample to 'H' (Hourly) and take the Mean
hourly_data = df.resample('H').mean()

print(hourly_data)
```
