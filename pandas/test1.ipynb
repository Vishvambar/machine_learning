```python
import pandas as pd
data = {
    "Product": ["Mouse", "Keyboard", "Laptop", "Monitor", "Headset", "Mousepad"],
    "Category": ["Accessory", "Accessory", "Device", "Device", "Accessory", "Accessory"],
    "Price": [20, 50, 1000, 200, 80, 10],
    "Sales_Qty": [100, 50, 10, 20, 30, 200]
}
df = pd.DataFrame(data)
print(df.groupby("Category")["Price"].mean())
```

```python
Revenue = df["Price"] * df["Sales_Qty"]
df["Revenue"] = Revenue
```

```python
df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
```
