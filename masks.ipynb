We have a stream of incoming patient vitals. we don't care about the healthy ones right now. We need to instantly isolate the Critical patients.

```python
import numpy as np 
```

```python
# Columns: [Patient_ID, Heart_Rate, Oxygen_Level]
# ID is just a number here for simplicity
patient_data = np.array([
    [101, 72, 98],   # Healthy
    [102, 115, 90],  # High HR, but Oxygen OK
    [103, 60, 99],   # Healthy
    [104, 120, 85],  # CRITICAL: High HR + Low Oxygen
    [105, 55, 88],   # CRITICAL: Low HR + Low Oxygen
    [106, 250, 98]   # SENSOR ERROR (Heart rate 250? Unlikely)
])
```

```python
patient_data
```

```python
oxygen_col =patient_data[: , 2  ]
```

```python
oxygen_col
```

```python
ishypoxic = oxygen_col < 90
```

```python
ishypoxic
```

```python
critical_patients = patient_data[ishypoxic]
critical_patients
```
