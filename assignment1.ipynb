You have a list of product prices: prices = np.array([100, 50, 200, 45, 300, 25, 600])

Task:

Calculate the mean of the prices.

Create a mask to find all prices that are below the mean.

Calculate the count of how many items are below the mean. (Hint: True is 1, False is 0).

Calculate the average price of just those "cheap" items.

```python
import numpy as np 
```

```python
prices = np.array([100, 50, 200, 45, 300, 25, 600])
```

```python
print(f"{prices.mean()}")
```

```python
below_mean_mask = prices < prices.mean()
```

```python
below_mean_mask
```

```python
print(f"below mean products count : {below_mean_mask.sum()}")
```

```python
cheap_items= prices[below_mean_mask]
```

```python
cheap_items
```

```python
average_chip_item = cheap_items.mean()
```

```python
print(average_chip_item)
```
