Grid pattern in broadcasting 

```python
import numpy as np 
import sys
import matplotlib.pyplot as plt 
np.set_printoptions(threshold=sys.maxsize)
```

```python
x = np.random.rand(100)
```

```python
y = np.random.rand(100)
```

```python
x =x.reshape(100,1)
```

```python
y =y.reshape(1, 100)
```

```python
m = x+y
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(np.broadcast_to(x, (100, 100)), cmap='viridis')
ax[0].set_title('X (Vertical Input)\nValues change Top-to-Bottom')
ax[1].imshow(np.broadcast_to(y, (100, 100)), cmap='viridis')
ax[1].set_title('Y (Horizontal Input)\nValues change Left-to-Right')
ax[2].imshow(x + y, cmap='viridis')
ax[2].set_title('X + Y\nThe Gradient Grid')
plt.tight_layout()
plt.show()
```
